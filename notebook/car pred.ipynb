```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
one=OneHotEncoder()
one.fit_transform(df1[['fueltype']])
```

```python
a=pd.DataFrame(data=one.fit_transform(df1[['fueltype']]).toarray(),columns=one.get_feature_names_out())
```

```python
one.fit_transform(df1[['aspiration']])
b=pd.DataFrame(data=one.fit_transform(df1[['aspiration']]).toarray(),columns=one.get_feature_names_out())
```

```python
one.fit_transform(df1[['doornumber']])
c=pd.DataFrame(data=one.fit_transform(df1[['doornumber']]).toarray(),columns=one.get_feature_names_out())
```

```python
oe=OrdinalEncoder(categories=[['convertible', 'hatchback', 'sedan', 'wagon', 'hardtop']])
df1['carbody']=oe.fit_transform(df1[['carbody']])
```

```python
one.fit_transform(df1[['drivewheel']])
d=pd.DataFrame(data=one.fit_transform(df1[['drivewheel']]).toarray(),columns=one.get_feature_names_out())
```

```python
one.fit_transform(df1[['enginelocation']])
e=pd.DataFrame(data=one.fit_transform(df1[['enginelocation']]).toarray(),columns=one.get_feature_names_out())
```

```python
oe=OrdinalEncoder(categories=[['dohc', 'ohcv', 'ohc', 'l', 'rotor', 'ohcf', 'dohcv']])
df1['enginetype']=oe.fit_transform(df1[['enginetype']])
```

```python
oe=OrdinalEncoder(categories=[['four', 'six', 'five', 'three', 'twelve', 'two', 'eight']])
df1['cylindernumber']=oe.fit_transform(df1[['cylindernumber']])
```

```python
oe=OrdinalEncoder(categories=[['mpfi', '2bbl', 'mfi', '1bbl', 'spfi', '4bbl', 'idi', 'spdi']])
df1['fuelsystem']=oe.fit_transform(df1[['fuelsystem']])
```

```python
df1=pd.concat([df1,a,b,c,d,e],axis=1)
```

```python
df1.head(2)
```

```python
df1.drop(['fueltype','aspiration','doornumber','drivewheel','enginelocation'],axis=1,inplace=True)
```

```python
df1.head(2)
```

```python
x=df1.drop(['price'],axis=1)
y=df1.price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,random_state=20,test_size=0.20)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train_ss=ss.fit_transform(X_train)
X_test_ss=ss.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_train_ss_pca=pca.fit_transform(X_train_ss)
X_test_ss_pca=pca.transform(X_test_ss)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_train_ss_pca[:,0],X_train_ss_pca[:,1])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
# from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import accuracy_score,r2_score
```

```python
model=[LinearRegression(),DecisionTreeRegressor(),SVR(),KNeighborsRegressor(),RandomForestRegressor(),GradientBoostingRegressor()]
```

```python
for i in model:
    p=i
    p.fit(X_train_ss,y_train)
    y_pred=p.predict(X_test_ss)
    #print(accuracy_score(y_pred,y_test))
    print(r2_score(y_pred,y_test))
    
```

```python
lr=GradientBoostingRegressor()
lr.fit(X_train_ss,y_train)
y_pred2=lr.predict(X_test_ss)
```

```python
print(r2_score(y_pred2,y_test))
```
